````python
#This solution is better because s and t consist of lowercase English letters only.

class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False

        char_count = [0] * 26

        for i in range(len(s)):
            char_count[ord(s[i]) - ord('a')] += 1;
            char_count[ord(t[i]) - ord('a')] -= 1;
            
        for x in char_count:
            if x !=0:
                return False;
        
        return True
    
    # Create an instance of Solution to call the method
sol = Solution()
print(sol.isAnagram("banana", "anabna")) # Output: True
print(sol.isAnagram("hello", "world"))    # Output: False
````

def two_sum(nums: List[int], target: int) -> List[int]:
    """
    Find two indices where nums[i] + nums[j] = target

    num_to_index = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_to_index:
            return [num_to_index[complement], i]
        num_to_index[num] = i
    
    return []

### Problem 1: Two Sum
- **üîó Link:** https://leetcode.com/problems/two-sum/
- **üìä Difficulty:** Easy

#### üìù Problem Description
```
Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

Example 1:
Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].
```

#### ü§î Initial Thoughts & Approach
- **First impression:** Need to find two numbers that add up to target
- **Key observations:** Only one solution exists, can't use same element twice
- **Edge cases to consider:** Empty array, array with one element, negative numbers
- **Approach:** Hash table to store complements for O(n) solution

#### üí° Solution Strategy
1. **Step 1:** Create a hash map to store number -> index mapping
2. **Step 2:** For each number, check if its complement (target - num) exists in map
3. **Step 3:** If complement exists, return both indices; otherwise, add current number to map

````python
# üîß Implementation for Two Sum
def two_sum(nums: List[int], target: int) -> List[int]:
    """
    Find two indices where nums[i] + nums[j] = target
    
    Args:
        nums: List of integers
        target: Target sum
    
    Returns:
        List of two indices
    """
    # Hash map to store number -> index
    num_to_index = {}
    
    for i, num in enumerate(nums):
        complement = target - num
        
        # Check if complement exists
        if complement in num_to_index:
            return [num_to_index[complement], i]
        
        # Store current number and its index
        num_to_index[num] = i
    
    # Should never reach here based on problem constraints
    return []

# Alternative brute force solution (for comparison)
def two_sum_brute_force(nums: List[int], target: int) -> List[int]:
    """Brute force approach - O(n¬≤) time complexity"""
    n = len(nums)
    for i in range(n):
        for j in range(i + 1, n):
            if nums[i] + nums[j] == target:
                return [i, j]
    return []
````

#### üìä Complexity Analysis
- **Time Complexity:** O(n) - Single pass through the array
- **Space Complexity:** O(n) - Hash map can store up to n elements
- **Explanation:** We traverse the array once, and hash table operations (insert/lookup) are O(1) on average

**Brute Force Comparison:**
- Time: O(n¬≤), Space: O(1) - but much slower for large inputs
